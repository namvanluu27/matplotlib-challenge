```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.DataFrame.merge(city_df, ride_df, how = 'outer', on = 'city')

# Display the data table for preview
pyber_df.head()
```

```python
# Your objective is to build a Bubble Plot that showcases the relationship between four key variables:
# Average Fare ($) Per City
# Total Number of Rides Per City
# Total Number of Drivers Per City
# City Type (Urban, Suburban, Rural)

# Create variables and establish values
city_group = pyber_df.groupby("city")
fare = city_group["fare"].mean()
total_rides = pyber_df["city"].value_counts()

# Cleaning drivers value and creating city type of Urban, Suburban, and Rural
clean_drivers_df = city_df.drop_duplicates(subset="city")
drivers_df = pd.DataFrame(clean_drivers_df)
drivers_df = drivers_df.set_index("city")
total_drivers = drivers_df["driver_count"]
city_type = drivers_df["type"]

# Creating Bubble Plot DataFrame and Displaying
bubble_dict={"Average Fare ($) Per City":fare,
      "Total Number of Rides Per City":total_rides,
       "Total Number of Drivers Per City":total_drivers,
       "City Type":city_type}

bubble_df=pd.DataFrame(bubble_dict)
bubble_df.head(40)

bubble_df.style.format("${:.2f}", subset = "Average Fare ($) Per City")
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Establish DataFrames for City Types
urban_df = bubble_df.loc[bubble_df["City Type"] == "Urban"]
suburban_df = bubble_df.loc[bubble_df["City Type"] == "Suburban"]
rural_df = bubble_df.loc[bubble_df["City Type"] == "Rural"]

# X values
urban_total_rides = urban_df["Total Number of Rides Per City"]
suburban_total_rides = suburban_df["Total Number of Rides Per City"]
rural_total_rides = rural_df["Total Number of Rides Per City"]

# Y values
urban_fare = urban_df["Average Fare ($) Per City"]
suburban_fare = suburban_df["Average Fare ($) Per City"]
rural_fare = rural_df["Average Fare ($) Per City"]

# Values for Circle Size
urban_driver = urban_df["Total Number of Drivers Per City"]
rural_driver = rural_df["Total Number of Drivers Per City"]
suburban_driver = suburban_df["Total Number of Drivers Per City"]


# Build the scatter plots for each city types
urban_plot = plt.scatter(urban_total_rides, urban_fare,s = urban_driver * 10, marker = "o", 
                       facecolors = "lightcoral",edgecolors = "black",  alpha = .7, linewidth = 1, label = "Urban")

suburban_plot = plt.scatter(suburban_total_rides, suburban_fare, suburban_driver * 10, marker = "o", 
                          facecolors = "lightskyblue", edgecolors = "black", alpha = .7, linewidth = 1, label = "Suburban")

rural_plot = plt.scatter(rural_total_rides, rural_fare, rural_driver * 10, marker = "o", 
                       facecolors = "gold",edgecolors = "black", alpha = .7, linewidth = 1, label = "Rural")

# Incorporate the other graph properties
plt.ylim(20,40)
plt.xlim(5,40)
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Rides (per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
plt.axis('auto')

# Create a legend
leg = plt.legend(title = "City Types", scatterpoints = 1)
leg.legendHandles[0]._sizes = [30]
leg.legendHandles[1]._sizes = [30]
leg.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42, 37, "Note:")
plt.text(42, 35 ,"Circle size correlates with driver count per city")

# Save Figure
plt.savefig("../Pyber/bubble_ride_shar.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fares = bubble_df["Average Fare ($) Per City"].sum()
total_fares

urban_total_fares = urban_df["Average Fare ($) Per City"].sum()
suburban_total_fares = suburban_df["Average Fare ($) Per City"].sum()
rural_total_fares = rural_df["Average Fare ($) Per City"].sum()

urban_fares_percent = (urban_total_fares / total_fares) * 100
suburban_fares_percent = (suburban_total_fares / total_fares) * 100
rural_fares_percent = (rural_total_fares / total_fares) * 100

percentage = {"Total Urban Fares Percent":urban_fares_percent,
              "Total Suburban Fares Percent": suburban_fares_percent,
              "Total Rural Fares Percent": rural_fares_percent}

# Build Pie Chart

labels = ["Urban", "Rural", "Suburban"]
sizes = [urban_fares_percent, rural_fares_percent, suburban_fares_percent]
colors = ["LightCoral", "Gold", "SkyBlue"]
explode = (0.1, 0, 0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors,
        autopct = "%1.1f%%", shadow = True, startangle = 260)

plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.legend(loc=4)

# Save Figure
plt.savefig("../Pyber/cityfarespercent.png")
```

```python
# Show Figure
plt.show()
```
